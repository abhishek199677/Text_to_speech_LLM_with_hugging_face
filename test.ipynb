```python
from transformers import pipeline
```

```python
text = "A powerful pipeline combining Large Language Models (LLMs) and neural TTS to generate natural-sounding speech from text. Built with Hugging Face transformers and optimized for realism, speed, and scalability"

# pipe = pipeline("text-to-speech", model="suno/bark-small")
pipe = pipeline("text-to-speech", model="suno/bark-small")

output = pipe(text)
```

```python
output
```

```python
from IPython.display import Audio

Audio(output["audio"], rate=output["sampling_rate"])
```
